```python
file2018 = open('Ei2018.pickle', 'rb')
Ei2018 = pickle.load(file2018)
file2018.close()
file2019 = open('Ei2019.pickle', 'rb')
Ei2019 = pickle.load(file2019)
file2019.close()
file2020 = open('Ei2020.pickle', 'rb')
Ei2020 = pickle.load(file2020)
file2020.close()
```

```python
file2018 = open('Wi2018.pickle', 'rb')
Wi2018 = pickle.load(file2018)
file2018.close()
file2019 = open('Wi2019.pickle', 'rb')
Wi2019 = pickle.load(file2019)
file2019.close()
file2020 = open('Wi2020.pickle', 'rb')
Wi2020 = pickle.load(file2020)
file2020.close()
```

```python
df_plottinglist2018_WE = pd.DataFrame(plottinglist2018_WE)
df_plottinglist2019_WE = pd.DataFrame(plottinglist2019_WE)
df_plottinglist2020_WE = pd.DataFrame(plottinglist2020_WE)
```

```python
df_plottinglist2018_WE.set_index('key', inplace=True)
df_plottinglist2019_WE.set_index('key', inplace=True)
df_plottinglist2020_WE.set_index('key', inplace=True)
```

```python
df_plottinglist2018_EN['ratioEbyN'] = df_plottinglist2018_EN['edges'] / df_plottinglist2018_EN['nodes']
df_plottinglist2019_EN['ratioEbyN'] = df_plottinglist2019_EN['edges'] / df_plottinglist2019_EN['nodes']
df_plottinglist2020_EN['ratioEbyN'] = df_plottinglist2020_EN['edges'] / df_plottinglist2020_EN['nodes']
df_plottinglist2018_WE['ratioEbyN'] = df_plottinglist2018_WE['weights'] / df_plottinglist2018_WE['edges']
df_plottinglist2019_WE['ratioEbyN'] = df_plottinglist2019_WE['weights'] / df_plottinglist2019_WE['edges']
df_plottinglist2020_WE['ratioEbyN'] = df_plottinglist2020_WE['weights'] / df_plottinglist2020_WE['edges']
```

```python
m_EN_2018 , b_EN_2018 = np.polyfit(df_plottinglist2018_EN.nodes.values, df_plottinglist2018_EN.edges.values, 1)
```

```python
m_EN_2018 , b_EN_2018 = np.polyfit(df_plottinglist2018_EN[df_plottinglist2018_EN.nodes<0.25].nodes.values, df_plottinglist2018_EN[df_plottinglist2018_EN.nodes<0.25].edges.values, 1)
m_EN_2019 , b_EN_2019 = np.polyfit(df_plottinglist2019_EN[df_plottinglist2019_EN.nodes<0.4].nodes.values, df_plottinglist2019_EN[df_plottinglist2019_EN.nodes<0.4].edges.values, 1)
m_EN_2020 , b_EN_2020 = np.polyfit(df_plottinglist2020_EN[df_plottinglist2020_EN.nodes<0.4].nodes.values, df_plottinglist2020_EN[df_plottinglist2020_EN.nodes<0.4].edges.values, 1)
m_WE_2018 , b_WE_2018 = np.polyfit(df_plottinglist2018_WE[df_plottinglist2018_WE.edges<0.8].edges.values, df_plottinglist2018_WE[df_plottinglist2018_WE.edges<0.8].weights.values, 1)
m_WE_2019 , b_WE_2019 = np.polyfit(df_plottinglist2019_WE[df_plottinglist2019_WE.edges<0.8].edges.values, df_plottinglist2019_WE[df_plottinglist2019_WE.edges<0.8].weights.values, 1)
m_WE_2020 , b_WE_2020 = np.polyfit(df_plottinglist2020_WE[df_plottinglist2020_WE.edges<0.8].edges.values, df_plottinglist2020_WE[df_plottinglist2020_WE.edges<0.8].weights.values, 1)
```

```python
outliercheck2018_EN = {}
for key in Ei2018:
    outliercheck2018_EN[key] = (max([math.log(Ei2018[key]),m_EN_2018*(math.log(Ni2018[key]))+b_EN_2018])/min([math.log(Ei2018[key]),m_EN_2018*(math.log(Ni2018[key]))+b_EN_2018])) * math.log(abs((math.log(Ei2018[key])-m_EN_2018*(math.log(Ni2018[key]))+b_EN_2018)/math.log(Ei2018[key]))+1)

outliercheck2019_EN = {}
for key in Ei2019:
    outliercheck2019_EN[key] = (max([math.log(Ei2019[key]),m_EN_2019*(math.log(Ni2019[key]))+b_EN_2019])/min([math.log(Ei2019[key]),m_EN_2019*(math.log(Ni2019[key]))+b_EN_2019])) * math.log(abs((math.log(Ei2019[key])-m_EN_2019*(math.log(Ni2019[key]))+b_EN_2019)/math.log(Ei2019[key]))+1)

outliercheck2020_EN = {}
for key in Ei2020:
    outliercheck2020_EN[key] = (max([math.log(Ei2020[key]),m_EN_2020*(math.log(Ni2020[key]))+b_EN_2020])/min([math.log(Ei2020[key]),m_EN_2020*(math.log(Ni2020[key]))+b_EN_2020])) * math.log(abs((math.log(Ei2020[key])-m_EN_2020*(math.log(Ni2020[key]))+b_EN_2020)/math.log(Ei2020[key]))+1)
```

```python
outliercheck2018_WE = {}
for key in Ei2018:
    outliercheck2018_WE[key] = (max([math.log(Ei2018[key]),m_WE_2018*(math.log(Ni2018[key]))+b_WE_2018])/min([math.log(Ei2018[key]),m_WE_2018*(math.log(Ni2018[key]))+b_WE_2018])) * math.log(abs((math.log(Ei2018[key])-m_WE_2018*(math.log(Ni2018[key]))+b_WE_2018)/math.log(Ei2018[key]))+1)

outliercheck2019_WE = {}
for key in Ei2019:
    outliercheck2019_WE[key] = (max([math.log(Ei2019[key]),m_WE_2019*(math.log(Ni2019[key]))+b_WE_2019])/min([math.log(Ei2019[key]),m_WE_2019*(math.log(Ni2019[key]))+b_WE_2019])) * math.log(abs((math.log(Ei2019[key])-m_WE_2019*(math.log(Ni2019[key]))+b_WE_2019)/math.log(Ei2019[key]))+1)

outliercheck2020_WE = {}
for key in Ei2020:
    outliercheck2020_WE[key] = (max([math.log(Ei2020[key]),m_WE_2020*(math.log(Ni2020[key]))+b_WE_2020])/min([math.log(Ei2020[key]),m_WE_2020*(math.log(Ni2020[key]))+b_WE_2020])) * math.log(abs((math.log(Ei2020[key])-m_WE_2020*(math.log(Ni2020[key]))+b_WE_2020)/math.log(Ei2020[key]))+1)
```

```python
# Plotting Values and Regression Line
max_x = np.max(df_plottinglist2018_EN.nodes.values)
min_x = np.min(df_plottinglist2018_EN.nodes.values)

# Calculating line values x and y
x = np.linspace(min_x, max_x, 1000)
y = b_EN_2018 + m_EN_2018 * x
 
# Ploting Line
plt.figure(figsize=(10,10))
plt.plot(x, y, color='#58b970', label='Regression Line')
# Ploting Scatter Points
plt.scatter(df_plottinglist2018_EN.nodes.values, df_plottinglist2018_EN.edges.values, c='#ef5423', label='Scatter Plot')
 
plt.xlabel('Log nodes')
plt.ylabel('Log edges')
plt.legend()

plt.show()
```

```python
# Plotting Values and Regression Line
max_x = np.max(df_plottinglist2020_WE.edges.values)
min_x = np.min(df_plottinglist2020_WE.edges.values)

# Calculating line values x and y
x = np.linspace(min_x, max_x, 1000)
y = b_WE_2020 + m_WE_2020 * x
 
# Ploting Line
plt.figure(figsize=(20,20))
plt.plot(x, y, color='b', label='Regression Line')
plt.plot(x, x, color='g', label='Slope=1 line',linestyle='dashed')
plt.plot(x, 2*x, color='r', label='Slope=2 line',linestyle='dashed')
# Ploting Scatter Points
plt.scatter(df_plottinglist2020_WE.edges.values, df_plottinglist2020_WE.weights.values, c='#ef5423', label='Scatter Plot')

# for key in df_plottinglist2018_WE.index:
#     plt.annotate(key, (df_plottinglist2018_WE.edges[key], df_plottinglist2018_WE.weights[key]))

df_plottinglist2020_WE_outlier_prediction_throughgraphs = []
for key in df_plottinglist2020_WE.index:
    if not ((df_plottinglist2020_WE.weights[key] > df_plottinglist2020_WE.edges[key]) and (df_plottinglist2020_WE.weights[key] < 2*df_plottinglist2020_WE.edges[key])):
        plt.annotate(key, (df_plottinglist2020_WE.edges[key], df_plottinglist2020_WE.weights[key]))
        df_plottinglist2020_WE_outlier_prediction_throughgraphs.append(key)
    
plt.xlabel('Log edges')
plt.ylabel('Log weights')
plt.legend()

plt.show()
```
